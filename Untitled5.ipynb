```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=[1,2,3]
y=[2,4,6]

plt.scatter(x,y)
plt.show()
```

```python
plt.plot(x,y)
plt.scatter(x,y)
plt.show()
```

```python
# x=np.array([1,2,3,4])
x=np.arange(0,5,0.1)
y=x **3
y2=x **2
plt.plot(x,y,label="x^3")
plt.plot(x,y2,label="x^2")
# plt.scatter(x,y
plt.ylabel("x^3")
plt.xlabel("x")
plt.grid()
plt.legend()
# plt.axis([0,10,0,200])
plt.title("plot")
plt.show()
```

```python
year=[2012,2013,2014,2015,2016,2017]
salary=[12,13,14,17,19,20]
population=[100,120,180,250,300,370]
plt.scatter(year,salary,s=population)
#bubble chart
plt.show()
```

```python
#pie graphs
labels=["class-A","Class-B","Class-C","Class-D"]
sizes=[3,4,6,2]
color=["blue","purple","red","pink"]
explode=[0.1,0,0,0]
plt.pie(sizes,colors=color,labels=labels,autopct="%.2f%%",explode=explode)
plt.axis("equal")

plt.title("Split among classes")
plt.show()
```

```python
xp=np.arange(6)
a=[1,1,2,2,3,3,3,4,5,5,5,5]
plt.hist(a,bins=4,edgecolor='black')
plt.xticks(xp)
plt.show()
```

```python
year=[2012,2013,2014,2015,2016,2017]
salary=[12,13,14,17,19,20]
population=[100,120,180,250,300,370]
plt.bar(year,population,width=0.6)
plt.xticks(rotation=40)
plt.show()
```
